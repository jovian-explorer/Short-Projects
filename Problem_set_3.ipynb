Using a program, preferably in python, plot $$\frac{2V_{eff}}{c^2}$$
vs. $$\frac{r}{(GM/c^2)} \hspace{2 mm} \text{or} \hspace{2 mm} \frac{r}{\mu}$$

for the Newtonian case, and for the GR cases for
$$h = \frac{J}{m} = 5.0 \mu c ,$$
$$h = \frac{J}{m} = 3.8 \mu c \text{ and }$$
$$h = \frac{J}{m} = 3.4 \mu c .$$

The $xâˆ’axis$ limits should be $0 \rightarrow 100.$. You can do this in a single plot, but 4 different plots are acceptable, so long as they have the same limits to facilitate comparison. Point out the differences between the Newtonian
and GR cases.

```python
import numpy as np
import matplotlib.pyplot as plt
from astropy import constants as con

#constants and variables
#c=speed of light
c = 1 
#a=mu*c^2=GM
a = 1
#r
r =  np.linspace(1,100,num=10000)
#x
x=r/(a/c**2)


#input the values of h 
h1= float(input("Enter the value of h1 in mpc: "))
h2= float(input("Enter the value of h2 in mpc: "))
h3= float(input("Enter the value of h3 in mpc: "))



#y equations for different h
#yn for relativistic case
#y_n for newtonian case

y1 = 2*((h1**2)/(2*(r **2 )) - a/r -(h1**2 * a)/(c**2 * r**3))/(c**2)
y_1= 2*((h1**2)/(2*r**2) - a/r)/c**2
y2 = 2*((h2**2)/(2*(r **2 )) - a/r -(h2**2 * a)/(c**2 * r**3))/c**2 
y_2= 2*((h2**2)/(2*r**2) - a/r)/c**2
y3 = 2*((h3**2)/(2*(r **2 )) - a/r -(h3**2 * a)/(c**2 * r**3))/c**2 
y_3= 2*((h3**2)/(2*r**2) - a/r)/c**2


fig, ax = plt.subplots()

ax_new = fig.add_axes([0.35, 0.35,0.35, 0.35])
ax_new.plot(x, y1, color='r', label='h1')
ax_new.plot(x, y2, color='g', label='h2')
ax_new.plot(x, y3, color='b', label='h3')

ax_new.plot(x, y_1, color='r', label='h1 N', linestyle='--')
ax_new.plot(x, y_2, color='g', label='h2 N', linestyle='--')
ax_new.plot(x, y_3, color='b', label='h3 N', linestyle='--')
ax_new.set_xlim([0, 11.5])
ax_new.set_ylim([-0.75, 0.5])

ax.plot(x, y1, color='r', label='h1')
ax.plot(x, y2, color='g', label='h2')
ax.plot(x, y3, color='b', label='h3')

ax.plot(x, y_1, color='r', label='h1 N', linestyle='--')
ax.plot(x, y_2, color='g', label='h2 N', linestyle='--')
ax.plot(x, y_3, color='b', label='h3 N', linestyle='--')
ax.set_ylim([-1, 1])


ax.legend()
```
