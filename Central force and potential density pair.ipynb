### (a) A particle moves in a potential $V (r) =- V_0 exp^{-\lambda^2r^2}$ , if the angular momentum is given by $L$ find the radius of  the circular orbit [Answer can be in the implicit form]. What is the largest value of $L$ for which a circular orbit exists and what is the value of effective potential $V_{eff}(r)$ at this critical orbit?

**Ans-**  $\hspace{2cm}$ The effective potential, $V_{eff} = \frac{L^2}{2mr^2} - V_0 exp^{-\lambda^2r^2}$   $\hspace{1cm}$---------------$(1)$


For stable circular orbit, $\frac{\partial V_{eff}(r)}{\partial r} = 0 $ $\hspace{7.0cm}$ ---------------$(2)$

Differentiating $(2)$ and equating to $(1)$, we have
$$\frac{-L^2}{mr^3} - V_0e^{-\lambda^2r^2}(-\lambda^2)(2r) = 0$$
$$\frac{L^2}{mr^3} = 2\lambda^2V_0re^{-\lambda^2r^2}$$
$$r^4e^{-\lambda^2r^2} = \frac{L^2}{2m\lambda^2 V_0}$$
$\therefore$ This is the implicit equation of r (radius of circular orbit)
$$L =\sqrt{2m\lambda^2 V_0 r^4 e^{-\lambda^2r^2}}$$

Here, for largest value of $L$, $\frac{\partial L}{\partial R} = 0$

 $\frac{\partial L}{\partial R}|_{r=r_{max}}=\frac{\sqrt{2m\lambda^2V_0}}{2\sqrt{r^4e^{-\lambda^2r^2}}}[r^4e^{-\lambda^2r^2}(-\lambda^2)2r + 4r^3e^{-\lambda^2r^2}] = 0$

$-2\lambda^2r^2e^{-\lambda^2r^2} +4e^{-\lambda^2r^2} =0$

$r^2 = \frac{2}{\lambda^2}$

$r_{max} = \frac{\sqrt{2}}{\lambda}$

Therefore, $L_{max} = \sqrt{2m\lambda^2V_0r_{max}^{4}e^{-\lambda^2r_{max}^2}}$

$\hspace{3.15cm}$ $= \sqrt{2m\lambda^2V_0}[(\frac{\sqrt{2}}{\lambda})^4e^{-\lambda^2(\frac{\sqrt{2}}{\lambda})^2}]^\frac{1}{2}$

$\hspace{1.9cm}$ $L_{max}=\frac{2}{\lambda e}\sqrt{2mV_0}$

and,$\hspace{1.05cm}$ $r_{max} = \frac{\sqrt{2}}{\lambda}$

Finally, using the values obtained in $(1)$, we have $V_{eff} =\frac{V_0}{e^2}$

```python
#importing libraries
import numpy as np
import matplotlib.pyplot as plt
#defining values
v_0 = 1
lam = 1
L = 1
m = 1
r = np.linspace(0,1,100)
def pot(r):
  return -v_0 * np.exp(-(lam**2 * r**2))

v = pot(r)
v_eff = (L**2 / (2 * m * r**2)) + v

# plt.plot(r,v, 'k--', label = 'r vs V')
plt.plot(r,v_eff,'r--', label = 'r vs $V_{eff}$')
plt.xlabel('radius $\longrightarrow$')
plt.ylabel('$ v_{eff} \longrightarrow$')
plt.legend()
plt.show()
```

### (c) Obtain the density distribution and circular velocity for a Jaffe potential given by - $$ùöΩ(r) = \frac{GM}{a} ln(\frac{r}{r+a})$$ where $M$ and $a$ are constants and show that at large distances the density corresponding the above potential goes as $r^{-4}$.

**Ans-**  In a spherical galaxy, the density of matter varies with radius as $\rho(r) = \frac{M}{4\pi}\frac{a}{r^2(r+a)^2}$, where $M$ and $a$ are constants.

Using Poisson‚Äôs Equation in spherical coordinates since the problem is spherically symmetric, only the $d/dr$ terms are relevant. 

$$\frac{1}{r^2}\frac{d}{dr}(r^2\frac{d\Phi}{dr}) = 4\pi G \rho$$
$$r^2\frac{d\Phi}{dr} = \int_0^r GMa\frac{dr}{(r+a)^2} = -GMa [\frac{1}{r+a}]^r_0 = GM\frac{r}{r+a}$$
$$\frac{d\Phi}{dr} = GM\frac{1}{r(r+a)} = \frac{GM}{a}(\frac{1}{r} - \frac{1}{r+a})$$
$$\Phi = \int^r_0 \frac{GM}{a}(\frac{1}{r} - \frac{1}{r+a})dr = \frac{GM}{a}(ln (r) - ln(r+a))dr = \frac{GM}{a} ln(\frac{r}{r+a})$$

```python
#defining constants
M=1
G=1
a=10

#defining density function
def density(G,a,r,M):
  return (M*a)/(4*np.pi*r**2*(r+a)**2)
#defining velocity function
def vel(G,a,r,M):
  return np.sqrt((G*M)/(r+a))

r = np.linspace(0.1,10,10000)
d = density(G,a,r,M)
#plotting
plt.plot(r,d,'r--',label = 'density')
plt.title("Density distribution for a Jaffe potential")
plt.xlabel('radius $\longrightarrow$')
plt.ylabel('Density $\longrightarrow$')
plt.yscale('log')
plt.legend()
```

```python
v = vel(G,a,r,M)
plt.plot(r,v,'g',label ='velocity')
#plotting
plt.xscale('log')
plt.title("Velocity distribution for a Jaffe potential")
plt.xlabel('radius $\longrightarrow$')
plt.ylabel('Velocity $\longrightarrow$')
plt.legend()
```

### **Further plot the dependence (in the same curve) of circular velocity with radial co-ordinate for cases (i) $r>>a$ and (ii) $a>>r$. Also compare the same with a standard Kepler profile.**

**Ans-** At large radii $r ‚â´ a, a/r ‚Üí 0$. Remember for small $x, ln(1 + x) ‚âà x$.
$$\Phi= \frac{GM}{a} ln(\frac{r}{r+a}) = \frac{GM}{a} ln(1 + \frac{a}{r})^{-1} = -  \frac{GM}{a} ln(1 + \frac{a}{r}) ‚âà -\frac{GM}{r}$$
which is the potential for a point mass.

$$v_c^2 = r|\frac{d\Phi}{dr}| = \frac{GM}{r+a}$$
For $r ‚â™ a$, this reduces to $v_c(r) \propto \frac{GM}{a}$, which is indpendent of r.

For $r‚â´ a$, this reduces to $v_c(r) \propto \frac{GM}{r}$, which is the Keplarian independence $v_c \propto r^{-1/2}.$

```python
G = 1
m = 1
a2 = 0.00001
a1 = 1000000*np.ones(100)
a2 = 0.00001
r = np.linspace(1, 10, 100)

def velocity(r, a):
  return (r*(G*m/a)*(1/r - 1/(r+a)))**0.5

v1 = velocity(r, a1)
v2 = velocity(r, a2)

plt.plot(r, 1/r**0.5,"r.-",label="Keplarian ")
plt.plot(r, v1, 'k-.', label="a>>r")
plt.plot(r, v2, 'g.-', label="a<<r")

plt.legend()
plt.show()
```
